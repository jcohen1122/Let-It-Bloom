# Clean Data

```python
#Get rid of null coordinates and reset index
df = finalData.dropna()
df = df.reset_index()
df = df[['Lat','Lon','Zone']]
#round to 2 decimal places
df = df.round(2)
df
```

Need to round up to xx.75 or down to xx.25 to match Koppen Geiger data

```python
#Round up to xx.75 or down to xx.25
import math
lat = []
lon = []

#Latitudes
for lt in df['Lat']:
  if abs(0.25 - (lt % 1)) > 0.25:
    lat.append(math.ceil(lt) - 0.25)
  else:
    lat.append(math.ceil(lt) - 0.75)

#Longitudes
for ln in df['Lon']:
  if abs(0.25 - (ln % 1)) > 0.25:
    lon.append(math.ceil(ln) - 0.25)
  else:
    lon.append(math.ceil(ln) - 0.75)

#Update dataframe lat and lon
df['Lat'] = lat
df['Lon'] = lon
```

```python
#sort data
df = pd.DataFrame.sort_values(by = ['Lat', 'Lon'], self = df)
df
```

```python
#Final CSV
df = df.to_csv('/content/drive/MyDrive/Tamid Group/Let it Bloom/Plant Hardiness Data/plant_hardiness.csv')
```
